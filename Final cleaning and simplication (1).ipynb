```python
breakdown={}
for k,v in majors_dict.items():
    percents=[]
    ones = sum(el in v for el in One)
    percents.append(ones)
    twos = sum(el in v for el in Two)
    percents.append(twos)
    threes= sum(el in v for el in Three)
    percents.append(threes)
    fours = sum(el in v for el in Four)
    percents.append(fours)
    breakdown.update({k:percents})
    
```

Classification breakdown for each major groups. graphs archived 

```python
import matplotlib.pyplot as plt
import numpy as np
import os
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)
def vizualize(major):
    counts = breakdown.get(major)
    names= 'Normal', 'Nine to Twelve', 'Thirteen Plus', 'Non Graduate'
    size = [counts[3], counts[2], counts[1], counts[0]]
    explode = [0,0,0,0.1]
    fig = plt.figure(dpi=150)
    fig.patch.set_facecolor('white')
    plt.rcParams['text.color'] = 'black'
    plt.pie(size, labels=names,autopct=lambda pct: func(pct, size), explode= explode)
    p=plt.gcf()
    output_path = os.path.join('graphs',major+'.png')
    plt.savefig(output_path)
    
```
